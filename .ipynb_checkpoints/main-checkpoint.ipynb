# DATA PREPARATION

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# setup
sns.set(style="darkgrid")
plt.rcParams["figure.figsize"] = (14, 6)
```

```python
train = pd.read_csv('data/train.csv', parse_dates=['date'])
test = pd.read_csv('data/test.csv', parse_dates=['date'])
stores = pd.read_csv('data/stores.csv')
stores = stores.rename(columns={'type': 'store_type', 'cluster': 'store_cluster'})

oil = pd.read_csv('data/oil.csv', parse_dates=['date'])
holidays = pd.read_csv('data/holidays_events.csv', parse_dates=['date'])
```

```python
display(train.tail(10))
```

```python
# combine test and training data
train['is_test'] = 0
test['is_test'] = 1
train = pd.concat([train, test], axis=0)
```

```python
# nulls
print("Missing values:\n", train.isnull().sum())
```

```python
# promotional effect

# Create bins for promotional count
bins = [-1, 0, 5, 10, 20, 50, 100, 5000]
labels = ['0', '1–5', '6–10', '11–20', '21–50', '51–100', '101+']
train['promo_bin'] = pd.cut(train['onpromotion'], bins=bins, labels=labels)

# Group by promo_bin and calculate average sales
promo_effect = train.groupby('promo_bin')['sales'].mean()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=promo_effect.index, y=promo_effect.values)
plt.title("Average Sales vs Number of Items on Promotion")
plt.xlabel("Number of Items on Promotion (Binned)")
plt.ylabel("Average Sales")
plt.tight_layout()
plt.show()
```

```python
# effect of holiday
# Keep only real holidays (not transferred) and drop duplicates on date
holidays = holidays[~holidays['transferred']].copy()
holidays = holidays[['date', 'locale']].drop_duplicates()

train = train.merge(holidays, on='date', how='left')

# Fill NaN with 'none' for non-holiday days
train['holiday_type'] = train['locale'].fillna('none')

# Drop intermediate column if you want
train = train.drop(columns=['locale'])

# Group by holiday type and calculate average sales
holiday_type_sales = train.groupby('holiday_type')['sales'].mean().reset_index()

# Sort categories logically
holiday_order = ['none', 'Local', 'Regional', 'National']
holiday_type_sales['holiday_type'] = pd.Categorical(holiday_type_sales['holiday_type'], categories=holiday_order, ordered=True)
holiday_type_sales = holiday_type_sales.sort_values('holiday_type')

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=holiday_type_sales, x='holiday_type', y='sales', palette='Set2')
plt.title("Average Sales by Holiday Type")
plt.xlabel("Holiday Type")
plt.ylabel("Average Sales")
plt.tight_layout()
plt.show()
```

```python
# effect of oil 
# Merge oil data with daily sales
daily_sales = train.groupby('date')['sales'].sum().reset_index()
merged = pd.merge(daily_sales, oil, on='date', how='left').dropna()

plt.figure(figsize=(70,40))
fig, ax1 = plt.subplots()
ax1.set_xlabel('Date')
ax1.set_ylabel('Sales', color='tab:blue')
ax1.plot(merged['date'], merged['sales'], color='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Oil Price', color='tab:red')
ax2.plot(merged['date'], merged['dcoilwtico'], color='tab:red')

plt.title("Sales vs Oil Price Over Time")
plt.show()

#find pearson's correlation coefficient
sales_oil = merged[['sales', 'dcoilwtico']]
sales_oil.corr()
```

# FEATURE ENGINEERING

```python
# findings: much stronger sales on the weekends, stronger sales very late/very early of the month
# store number, store types, store cluster are important signals
# product type, promotion crucially important
# holiday vs non-holiday is important
```

```python
# weekday/weekend features
train['is_weekend'] = train['day_of_week'].isin([5, 6]).astype(int)

# time-of-month segments
def time_of_month(day):
    if day <= 5 or day >= 30:
        return 1
    else:
        return 0
        
train['early_late_flag'] = train['day_of_month'].apply(time_of_month)

# month of year (flag december)
train['is_december'] = (train['date'].dt.month == 12).astype(int)

# store type
train = train.merge(stores[['store_nbr', 'store_type']], on='store_nbr', how='left')

# store cluster
train = train.merge(stores[['store_nbr', 'store_cluster']], on='store_nbr', how='left')
```

```python
# ENCODING
from sklearn.preprocessing import LabelEncoder

columns_to_label_encode = ['promo_bin', 'holiday_type']
encoders = {}

for col in columns_to_label_encode:
    le = LabelEncoder()
    le.fit(train[train['is_test'] == 0][col])
    train[col] = le.transform(train[col])
    encoders[col] = le
```

```python
# features I still need to add
# oil price, oil lag features, oil rolling promotion windows
# store family interaction

# train_raw = train.copy()
# display(train.tail())
```

```python
# lag features
train = train.sort_values(by=['store_nbr', 'family', 'date'])

# Create lag features
train['sales_lag_1'] = train.groupby(['store_nbr', 'family'])['sales'].shift(1)
train['sales_lag_7'] = train.groupby(['store_nbr', 'family'])['sales'].shift(7)
train['sales_lag_14'] = train.groupby(['store_nbr', 'family'])['sales'].shift(14)
train['sales_lag_28'] = train.groupby(['store_nbr', 'family'])['sales'].shift(28)
```

```python
# Rolling 7-day mean (shifted to avoid leakage)
train['rolling_mean_3'] = train.groupby(['store_nbr', 'family'])['sales'].shift(1).rolling(window=3).mean()
train['rolling_mean_7'] = train.groupby(['store_nbr', 'family'])['sales'].shift(1).rolling(window=7).mean()
train['rolling_mean_14'] = train.groupby(['store_nbr', 'family'])['sales'].shift(1).rolling(window=14).mean()

# Rolling 7-day std dev
train['rolling_std_3'] = train.groupby(['store_nbr', 'family'])['sales'].shift(1).rolling(window=3).std()
train['rolling_std_7'] = train.groupby(['store_nbr', 'family'])['sales'].shift(1).rolling(window=7).std()
train['rolling_std_14'] = train.groupby(['store_nbr', 'family'])['sales'].shift(1).rolling(window=14).std()
```

```python
# Rolling 7-day promo count
train['promo_rolling_7'] = (
    train.groupby(['store_nbr', 'family'])['onpromotion']
    .shift(1)  # Shift to prevent leakage
    .rolling(window=7)
    .sum()
)

# Rolling 14-day promo count
train['promo_rolling_14'] = (
    train.groupby(['store_nbr', 'family'])['onpromotion']
    .shift(1)
    .rolling(window=14)
    .sum()
)

train = train.dropna(subset=['promo_rolling_7', 'promo_rolling_14'])
```

```python
# oil features

train = train.merge(oil[['date', 'dcoilwtico']], on='date', how='left')
train = train.rename(columns={'dcoilwtico': 'oil'})

train['oil_lag_7'] = train.groupby(['store_nbr', 'family'])['oil'].shift(7)
train['oil_lag_30'] = train.groupby(['store_nbr', 'family'])['oil'].shift(30)
train['oil_lag_60'] = train.groupby(['store_nbr', 'family'])['oil'].shift(60)


train = train.drop(columns=['oil'])
```

```python
# data-time features
train['week_of_year'] = train['date'].dt.isocalendar().week

#cyclic features
train['dow_sin'] = np.sin(2 * np.pi * train['day_of_week'] / 7)
train['dow_cos'] = np.cos(2 * np.pi * train['day_of_week'] / 7)

train['woy_sin'] = np.sin(2 * np.pi * train['week_of_year'] / 52)
train['woy_cos'] = np.cos(2 * np.pi * train['week_of_year'] / 52)

train['month_sin'] = np.sin(2 * np.pi * train['month'] / 12)
train['month_cos'] = np.cos(2 * np.pi * train['month'] / 12)

# was it payday in the past three days?
paydays = [15, 30, 31]
train['is_payday'] = train['date'].dt.day.isin(paydays).astype(int)
train['is_payday_lag_3'] = (
    train
    .groupby(['store_nbr', 'family'], group_keys=False)['is_payday']
    .apply(lambda x: x.shift(1).rolling(3, min_periods=1).max())
    .fillna(0)
    .astype(int)
)
```

## Train-test-validation split & final data preparation

```python
# check for null values here
print(train.isnull().sum().sort_values(ascending=False).head(20))
```

```python
test = train[train['is_test'] == 1].copy()
train = train[train['is_test'] == 0].copy()
```

```python
display(train.info())
```

```python
train.drop(columns=['is_test'], inplace=True)
test.drop(columns=['is_test'], inplace=True)
```

```python
#different training set for linear regression baseline bc it can't handle null or string values
lin_train = train.copy()
lin_test = test.copy()

required_features = [
    'oil_lag_60', 'oil_lag_30', 'oil_lag_7',
    'sales_lag_1', 'sales_lag_7', 'sales_lag_14', 'sales_lag_28',
    'rolling_mean_3', 'rolling_mean_7', 'rolling_mean_14',
    'rolling_std_3', 'rolling_std_7', 'rolling_std_14'
]

lin_train = lin_train.dropna(subset=required_features)

lin_train.drop(columns=['store_type', 'family'], inplace=True)


# Define target and features
target = 'sales'
drop_cols = ['id','sales']

features = [col for col in lin_train.columns if col not in drop_cols]

X_lin_test = lin_test[features].copy()
y_lin_test = lin_test[target].copy()

val_start_date = '2017-08-01'

X_lin_train = lin_train[lin_train['date'] < val_start_date][features].copy()
y_lin_train = lin_train[lin_train['date'] < val_start_date][target].copy()

X_lin_val = lin_train[lin_train['date'] >= val_start_date][features].copy()
y_lin_val = lin_train[lin_train['date'] >= val_start_date][target].copy()

X_lin_train.drop(columns =['date'], inplace=True)
X_lin_val.drop(columns =['date'], inplace=True)
X_lin_test.drop(columns =['date'], inplace=True)
```

```python
#catBoost training set
id_df = X_test.pop('id')

X_test = test[features].copy()

X_train = train[train['date'] < val_start_date][features].copy()
y_train = train[train['date'] < val_start_date][target].copy()

X_val = train[train['date'] >= val_start_date][features].copy()
y_val = train[train['date'] >= val_start_date][target].copy()

X_train.drop(columns =['date'], inplace=True)
X_val.drop(columns =['date'], inplace=True)
X_test.drop(columns =['date'], inplace=True)
```

# MODEL TRAINING

## LINEAR REGRESSION = BASELINE

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_log_error

y_train_log = np.log1p(y_lin_train)
y_val_log = np.log1p(y_lin_val)

# Train the model
linreg = LinearRegression()
linreg.fit(X_lin_train, y_train_log)  # train on log-transformed sales

# Predict
y_pred_log = linreg.predict(X_lin_val)
y_pred = np.expm1(y_pred_log)  # back to original scale

# Evaluate
rmsle = np.sqrt(root_mean_squared_log_error(y_lin_val, y_pred.clip(0, None)))
print(f"Validation RMSLE (Linear Regression): {rmsle:.4f}")
```

## catBOOOOOOST

```python
from catboost import CatBoostRegressor
from sklearn.metrics import root_mean_squared_log_error
import numpy as np

# identify cateogircal features
categorical_features = ['store_type', 'family']

cat_model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.05,
    depth=6,
    loss_function='RMSE',
    random_seed=42,
    verbose=100,
    cat_features=categorical_features
)

# log-transform sales
cat_model.fit(
    X_train,
    np.log1p(y_train),
    eval_set=(X_val, np.log1p(y_val)),
    use_best_model=True
)

# predict and invert log transform
y_pred_log = cat_model.predict(X_val)
y_pred = np.expm1(y_pred_log)

# Compute RMSLE
rmsle = np.sqrt(root_mean_squared_log_error(y_val, y_pred.clip(0, None)))
print(f"Validation RMSLE (CatBoost): {rmsle:.4f}")
```

## CHECK FEATURE IMPORTANCE

```python
feature_importances = cat_model.get_feature_importance()
feature_names = cat_model.feature_names_

# Combine into DataFrame
fi_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False) 

top_n = 30  # or 39 if you want to match your old plot
top_features = fi_df.head(top_n)

plt.figure(figsize=(10, 8))
plt.barh(top_features['Feature'], top_features['Importance'])
plt.title("Top CatBoost Features by Importance")
plt.xlabel("Importance Score")
plt.gca().invert_yaxis()  # highest importance at the top
plt.tight_layout()
plt.show()
```
